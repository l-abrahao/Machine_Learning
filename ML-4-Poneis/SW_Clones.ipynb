## Preparando os Dados

```python
features = ["Estatura(cm)",	"Massa(em kilos)", "Distância Ombro a ombro",	"Tamanho do crânio",	"Tamanho dos pés"]

cat_features = ["Distância Ombro a ombro","Tamanho do crânio","Tamanho dos pés"]
x = df[features]
x
```

```python
!pip install feature-engine
```

```python
from feature_engine import encoding
```

```python
onehot = encoding.OneHotEncoder(variables = cat_features)
onehot.fit(x)

x = onehot.transform(x)
x
```

## Modelo

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
arvore.fit(x,df['Status_bool'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(dpi=250)

class_names_str = [str(cls) for cls in arvore.classes_]
tree.plot_tree(arvore, class_names=class_names_str, feature_names = x.columns, filled=True, max_depth=4)
```
